```python
imgHSV = cv.cvtColor(image, cv.COLOR_BGR2HSV)

lower_bound_green = np.array([35,50,50])
upper_bound_green = np.array([85,255,255])

lower_bound_blue = np.array([100,50,50])
upper_bound_blue = np.array([130,255,255])

lower_bound_red1 = np.array([0, 50, 50])
upper_bound_red1 = np.array([10, 255, 255])

lower_bound_red2 = np.array([170, 50, 50])
upper_bound_red2 = np.array([180, 255, 255])

mask_green = cv.inRange(imgHSV, lower_bound_green, upper_bound_green)
segmented_green = cv.bitwise_and(image, image, mask=mask_green)
segmented_green_rgb = cv.cvtColor(segmented_green, cv.COLOR_BGR2RGB)

mask_blue = cv.inRange(imgHSV, lower_bound_blue, upper_bound_blue)
segmented_blue = cv.bitwise_and(image, image, mask=mask_blue)
segmented_blue_rgb = cv.cvtColor(segmented_blue, cv.COLOR_BGR2RGB)

mask_red1 = cv.inRange(imgHSV, lower_bound_red1, upper_bound_red1)
mask_red2 = cv.inRange(imgHSV, lower_bound_red2, upper_bound_red2)
mask_red = mask_red1 | mask_red2
segmented_red = cv.bitwise_and(image, image, mask=mask_red)
segmented_red_rgb = cv.cvtColor(segmented_red, cv.COLOR_BGR2RGB)


plt.figure(figsize=(12,18))
# plt.subplot(1,1,1)
# plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))

# GREEN MASKED

plt.subplot(4,2,1)
plt.imshow(mask_green, cmap='gray')
plt.axis("off")

plt.subplot(4,2,2)
plt.imshow(segmented_green_rgb, cmap='hsv')
plt.axis("off")

# BLUE MASKED

plt.subplot(4,2,3)
plt.imshow(mask_blue, cmap='gray')
plt.axis("off")

plt.subplot(4,2,4)
plt.imshow(segmented_blue_rgb)
plt.axis("off")

# RED MASKED

plt.subplot(4,2,5)
plt.imshow(mask_red1, cmap='gray')
plt.axis("off")

plt.subplot(4,2,6)
plt.imshow(segmented_red_rgb)
plt.axis("off")

# COMBINING
combine_rg = cv.bitwise_or(segmented_red, segmented_green)
combine_rgb = cv.bitwise_or(combine_rg, segmented_blue)

plt.figure(figsize=(12,4))
plt.subplot(1,1,1)
plt.imshow(cv.cvtColor(combine_rgb, cv.COLOR_BGR2RGB) )
plt.axis("off")

plt.show()
```
